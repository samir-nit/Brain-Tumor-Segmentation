```python
import pandas as pd
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips  #dataframe 
```

```python
sns.set() #sets all default styles for plotting graphs
```

```python
sns.distplot(tips['total_bill'], kde = False) #kernel density estimate 
```

```python
sns.distplot(tips['total_bill'], hist = False) #kernel density estimate 
```

```python
sns.scatterplot(x="total_bill",y="tip",data = tips, alpha=0.8, s=100, hue ='day')
```

```python
sns.relplot(x="total_bill", y = "day" , data = tips, hue = "size", style = "smoker") #relational plot 
```

```python
sns.jointplot(x="total_bill", y = "tip", data = tips, kind='resid') #residual plot
sns.jointplot(x="total_bill", y = "tip", data = tips, kind='scatter') 
sns.jointplot(x="total_bill", y = "tip", data = tips, kind='reg', color = 'cyan')  
sns.jointplot(x="total_bill", y = "tip", data = tips, kind='kde', color ='purple')
sns.jointplot(x="total_bill", y = "tip", data = tips, kind='hex', color = 'red')
```

```python
sns.pairplot(tips)
```

```python
sns.stripplot(x="day", y = "total_bill", data = tips)
```

```python
sns.swarmplot(x="day", y = "total_bill", data = tips)
```

```python
sns.boxplot(x="day", y="total_bill", data = tips)
#sns.boxplot(x=tips["total_bill"]) #for only total_bill
```

```python
sns.violinplot(x=tips["total_bill"])
```

```python
#categorical plots
sns.catplot(x='day',y='total_bill',data=tips, kind = 'violin')
sns.catplot(x='day',y='total_bill',data=tips, kind = 'swarm')
sns.catplot(x='day',y='total_bill',data=tips, kind = 'box')
sns.catplot(x='day',y='total_bill',data=tips, kind = 'strip')
sns.catplot(x='day',y='total_bill',data=tips, kind = 'boxen')
```
